```python
import heapq
```

```python
def assignTasks(servers, tasks, ref = []):
    # tasks seconds
    # servers weights

    free_servers = [(weight, index) for index,weight in enumerate(servers)]
    heapq.heapify(free_servers)

    processing_servers = []  # end_time, index, weight
    heapq.heapify(processing_servers)

    time = 0
    res = [-1]*len(tasks)

    for task_idx, proc_time in enumerate(tasks):
        time = max(time, task_idx)
        if not free_servers:
            end_time, index, weight = heapq.heappop(processing_servers)
            heapq.heappush(free_servers, (weight, index))
            time = max(time, end_time)
        while processing_servers and processing_servers[0][0] <= time:
            end_time, index, weight = heapq.heappop(processing_servers)
            heapq.heappush(free_servers, (weight, index))

        weight, index = heapq.heappop(free_servers)
        heapq.heappush(processing_servers, (time+proc_time, index, weight))
        res[task_idx] = index
        
        if ref:
            if res[task_idx] != ref[task_idx]:
                assert False
            print(processing_servers)
            print(free_servers)
            print()

    return res
```
